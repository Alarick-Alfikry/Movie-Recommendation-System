**<h3>Movie Recommendation System Using ✨Python✨**

**<p1>Import Libraries**

```python
# HAHAA IMPORT GO BRRRRR 

import numpy as np 
import pandas as pd 
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import HTML
import ipywidgets as widgets
```

**<p1>Colecting and proses the data**

```python
# load n check movie data

pd.set_option('display.max_columns', 100)

movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies.head()
```

```python
credits.head()
```

```python
# combine 2 columns

merged_df = pd.merge(movies, credits, left_on='id', right_on='movie_id', how='inner', suffixes=('_movies', '_credits'))

merged_df.head()
```

```python
# select important columns

select_column = ['title_movies', 'genres', 'tagline', 'keywords', 'cast', 'production_companies']
```

```python
# fill null values with null strings

for x in select_column : 
    merged_df[x] = merged_df[x].fillna('')
```

```python
# merge columns 

combine_column = merged_df['title_movies'] + ' ' + merged_df['tagline'] + ' ' + merged_df['keywords']  + ' ' + merged_df['cast']  + ' ' + merged_df['production_companies'] + ' ' + merged_df['genres']
```

```python
# convert text to numerical (feature vector)

vectorizer = TfidfVectorizer()

vector = vectorizer.fit_transform(combine_column)

print(vector[:5])
```

**<p1>Getting Cosine Similarity**

```python
# find similarity points

similarity = cosine_similarity(vector)

print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user

movie_input = input('Enter your favourite movie name : ')  # e.g : batman
```

```python
# creating list of movie

movie_list =  merged_df['title_movies'].tolist()
print(movie_list[:10])
```

```python
# finding the closes match of movie and index from movie input


movie_match = difflib.get_close_matches(movie_input, merged_df['title_movies'], n=1)

if movie_match:
    close_movie_title = movie_match[0]
    movie_index = (merged_df['title_movies'] == close_movie_title).idxmax()

    print("Closest match:", close_movie_title)
    print("Index of closest match:", movie_index)
else:
    print("No close match found.")
```

```python
# getting similar movie

match_score = list(enumerate(similarity[movie_index]))

print(match_score[:5])
```

```python
# sorting the movie based on match score 

sorted_match_movie = sorted(match_score, key= lambda x:x[1], reverse=True)

print(sorted_match_movie[:10])
```

```python
# print the name of similar movies based on index

print("Movie for u : \n")

i = 1 
for movie in sorted_match_movie: 
    index = movie[0]
    title_Findex = merged_df.loc[index, 'title_movies']
    if i < 16: 
        print(i, '.', title_Findex)
        i += 1
```

**<h3>Movie recommendation system**

<center><h3><b>Movie Recommendation System</b></h3></center>
<center><p>Movie Recommendation System using ✨Python✨ Discover your next movie night based on your favorite genres and preferences!</p></center>

```python
from IPython.display import display

# function to perform movie search and display results
def search_and_display(change):
    movie_input_value = change['new']
    
    if not movie_input_value:
        # If input is empty, do nothing
        return
    
    movie_match = difflib.get_close_matches(movie_input_value, merged_df['title_movies'], n=1)

    movie_list.clear_output(wait=True)  # clear previous output

    if movie_match:
        close_movie_title = movie_match[0]
        movie_index = (merged_df['title_movies'] == close_movie_title).idxmax()

        with movie_list:
            match_score = list(enumerate(similarity[movie_index]))
            sorted_match_movie = sorted(match_score, key=lambda x: x[1], reverse=True)

            print("Movies for you : \n")
            i = 1
            for movie in sorted_match_movie:
                index = movie[0]
                title_Findex = merged_df.loc[index, 'title_movies']
                if i < 16:
                    print(i, '.', title_Findex)
                    i += 1
    else:
        # do nothing if no match is found
        pass

# create input widget
movie_input = widgets.Text(
    description="Movie Title :",
    placeholder="Type ur fav movie here!",
    disabled=False,
    layout=widgets.Layout(width='50%')
)

# create output widget
movie_list = widgets.Output()

movie_input.observe(search_and_display, names='value')
display(movie_input, movie_list)
```
